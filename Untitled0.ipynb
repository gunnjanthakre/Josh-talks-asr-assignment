```python
# Cell: Full Whisper Fine-tuning Pipeline
import os, gc, torch
import pandas as pd
import numpy as np
import requests
import soundfile as sf
import librosa
from io import BytesIO
from datasets import Dataset, Audio
from transformers import (
    WhisperFeatureExtractor,
    WhisperTokenizer,
    WhisperProcessor,
    WhisperForConditionalGeneration,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer
)
import evaluate
from dataclasses import dataclass
from typing import Any, Dict, List, Union

print("✅ Imports done")
print(f"GPU: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'NO GPU!'}")

# ── 1. Load segments ──────────────────────────────────────────────────────────
seg_df = pd.read_csv('/content/drive/MyDrive/josh_talks_asr/all_segments.csv')
print(f"Total segments: {len(seg_df)}")

# ── 2. Load Whisper ───────────────────────────────────────────────────────────
MODEL = "openai/whisper-small"
feature_extractor = WhisperFeatureExtractor.from_pretrained(MODEL)
tokenizer = WhisperTokenizer.from_pretrained(MODEL, language="hindi", task="transcribe")
processor = WhisperProcessor.from_pretrained(MODEL, language="hindi", task="transcribe")
model = WhisperForConditionalGeneration.from_pretrained(MODEL)
model.generation_config.language = "hindi"
model.generation_config.task = "transcribe"
model.generation_config.forced_decoder_ids = None
print("✅ Whisper-small loaded")

# ── 3. Download audio segments ────────────────────────────────────────────────
def download_audio_segment(audio_url, start, end, target_sr=16000):
    """Download full audio and extract segment by timestamp."""
    try:
        r = requests.get(audio_url, timeout=30)
        if r.status_code != 200:
            return None
        audio_bytes = BytesIO(r.content)
        audio, sr = sf.read(audio_bytes)
        # Convert to mono
        if len(audio.shape) > 1:
            audio = audio.mean(axis=1)
        # Resample to 16kHz
        if sr != target_sr:
            audio = librosa.resample(audio, orig_sr=sr, target_sr=target_sr)
        # Extract segment
        start_sample = int(start * target_sr)
        end_sample = int(end * target_sr)
        segment = audio[start_sample:end_sample]
        return segment.astype(np.float32)
    except:
        return None

# Cache audio files to avoid re-downloading
audio_cache = {}

def get_segment(row):
    key = row['audio_url']
    if key not in audio_cache:
        try:
            r = requests.get(row['audio_url'], timeout=30)
            audio_bytes = BytesIO(r.content)
            audio, sr = sf.read(audio_bytes)
            if len(audio.shape) > 1:
                audio = audio.mean(axis=1)
            if sr != 16000:
                audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)
            audio_cache[key] = (audio.astype(np.float32), 16000)
        except:
            audio_cache[key] = None

    cached = audio_cache[key]
    if cached is None:
        return None
    audio, sr = cached
    start_s = int(row['start'] * sr)
    end_s = int(row['end'] * sr)
    return audio[start_s:end_s]

# ── 4. Build HuggingFace Dataset ──────────────────────────────────────────────
print("Building dataset (downloading audio)...")
print("This will take ~15-20 minutes...")

# Use first 80% for train, 20% for eval
# Group by recording to avoid data leakage
recordings = seg_df['recording_id'].unique()
np.random.seed(42)
np.random.shuffle(recordings)
split = int(0.8 * len(recordings))
train_recs = set(recordings[:split])
eval_recs = set(recordings[split:])

train_df = seg_df[seg_df['recording_id'].isin(train_recs)].reset_index(drop=True)
eval_df = seg_df[seg_df['recording_id'].isin(eval_recs)].reset_index(drop=True)

print(f"Train segments: {len(train_df)}")
print(f"Eval segments: {len(eval_df)}")

def prepare_batch(batch_df, name=""):
    data = []
    # Download unique audio files first
    unique_urls = batch_df['audio_url'].unique()
    print(f"Downloading {len(unique_urls)} audio files for {name}...")
    for url in unique_urls:
        dummy_row = batch_df[batch_df['audio_url']==url].iloc[0]
        get_segment(dummy_row)  # populate cache

    print(f"Extracting segments for {name}...")
    for _, row in batch_df.iterrows():
        seg = get_segment(row)
        if seg is not None and len(seg) > 1600:  # min 0.1s
            data.append({
                'audio': seg,
                'text': row['text']
            })
    return data

train_data = prepare_batch(train_df, "train")
eval_data = prepare_batch(eval_df, "eval")

print(f"✅ Train samples ready: {len(train_data)}")
print(f"✅ Eval samples ready: {len(eval_data)}")

# ── 5. Preprocessing ──────────────────────────────────────────────────────────
def preprocess(examples):
    audio = examples['audio']
    inputs = feature_extractor(audio, sampling_rate=16000, return_tensors='pt')
    input_features = inputs.input_features[0]
    labels = tokenizer(examples['text']).input_ids
    return {'input_features': input_features, 'labels': labels}

print("Preprocessing train set...")
train_processed = [preprocess(d) for d in train_data]
print("Preprocessing eval set...")
eval_processed = [preprocess(d) for d in eval_data]

from datasets import Dataset
train_dataset = Dataset.from_list(train_processed)
eval_dataset = Dataset.from_list(eval_processed)
print(f"✅ Datasets ready: train={len(train_dataset)}, eval={len(eval_dataset)}")

# ── 6. Data Collator ──────────────────────────────────────────────────────────
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any
    def __call__(self, features):
        input_features = [{"input_features": f["input_features"]} for f in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")
        label_features = [{"input_ids": f["labels"]} for f in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")
        labels = labels_batch["input_ids"].masked_fill(
            labels_batch.attention_mask.ne(1), -100)
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]
        batch["labels"] = labels
        return batch

data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)

# ── 7. Metrics ────────────────────────────────────────────────────────────────
wer_metric = evaluate.load("wer")

def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids
    label_ids[label_ids == -100] = tokenizer.pad_token_id
    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = tokenizer.batch_decode(label_ids, skip_special_tokens=True)
    wer = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"wer": wer}

# ── 8. Training ───────────────────────────────────────────────────────────────
save_path = '/content/drive/MyDrive/josh_talks_asr/whisper-small-hindi'

training_args = Seq2SeqTrainingArguments(
    output_dir=save_path,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2,
    learning_rate=1e-5,
    warmup_steps=100,
    max_steps=500,
    gradient_checkpointing=True,
    fp16=True,
    eval_strategy="steps",
    eval_steps=100,
    save_steps=100,
    logging_steps=25,
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False,
    predict_with_generate=True,
    generation_max_length=225,
    save_total_limit=2,
    report_to=["tensorboard"],
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    tokenizer=processor.feature_extractor,
)

print("🚀 Starting training...")
print("This takes ~2-3 hours. Do NOT close Colab tab!")
trainer.train()

# ── 9. Save ───────────────────────────────────────────────────────────────────
model.save_pretrained(save_path)
processor.save_pretrained(save_path)
print(f"✅ Model saved to {save_path}")
```

```python
# FIXED: Memory-efficient approach
import os, gc, torch
import pandas as pd
import numpy as np
import requests
import soundfile as sf
import librosa
from io import BytesIO

print(f"GPU: {torch.cuda.get_device_name(0)}")

# ── 1. Load segments ──────────────────────────────────────────────────────────
seg_df = pd.read_csv('/content/drive/MyDrive/josh_talks_asr/all_segments.csv')

# Use only 2000 segments to avoid RAM crash
seg_df = seg_df.sample(2000, random_state=42).reset_index(drop=True)
print(f"Using {len(seg_df)} segments")

# ── 2. Load Whisper ───────────────────────────────────────────────────────────
from transformers import WhisperFeatureExtractor, WhisperTokenizer, WhisperProcessor, WhisperForConditionalGeneration
MODEL = "openai/whisper-small"
feature_extractor = WhisperFeatureExtractor.from_pretrained(MODEL)
tokenizer = WhisperTokenizer.from_pretrained(MODEL, language="hindi", task="transcribe")
processor = WhisperProcessor.from_pretrained(MODEL, language="hindi", task="transcribe")
model = WhisperForConditionalGeneration.from_pretrained(MODEL)
model.generation_config.language = "hindi"
model.generation_config.task = "transcribe"
model.generation_config.forced_decoder_ids = None
print("✅ Whisper loaded")

# ── 3. Process ONE recording at a time, save features to disk ─────────────────
import pickle

save_dir = '/content/drive/MyDrive/josh_talks_asr/features'
os.makedirs(save_dir, exist_ok=True)

train_features = []
eval_features = []

# Split by recording
recordings = seg_df['recording_id'].unique()
np.random.seed(42)
np.random.shuffle(recordings)
split = int(0.8 * len(recordings))
train_recs = set(recordings[:split])
eval_recs = set(recordings[split:])

print(f"Train recordings: {len(train_recs)}, Eval recordings: {len(eval_recs)}")

processed = 0
failed = 0

for url, group in seg_df.groupby('audio_url'):
    try:
        # Download once per audio file
        r = requests.get(url, timeout=30)
        if r.status_code != 200:
            failed += len(group)
            continue

        audio, sr = sf.read(BytesIO(r.content))
        if len(audio.shape) > 1:
            audio = audio.mean(axis=1)
        if sr != 16000:
            audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)
        audio = audio.astype(np.float32)

        # Process each segment from this file
        for _, row in group.iterrows():
            start_s = int(row['start'] * 16000)
            end_s = int(row['end'] * 16000)
            seg = audio[start_s:end_s]

            if len(seg) < 1600:
                continue

            # Extract features immediately — don't store raw audio
            inputs = feature_extractor(seg, sampling_rate=16000, return_tensors='pt')
            input_features = inputs.input_features[0].numpy()
            labels = tokenizer(row['text']).input_ids

            sample = {
                'input_features': input_features,
                'labels': labels
            }

            if row['recording_id'] in train_recs:
                train_features.append(sample)
            else:
                eval_features.append(sample)

            processed += 1

        # Free audio from memory immediately
        del audio, r
        gc.collect()

        if processed % 100 == 0:
            print(f"  Processed {processed} segments...")

    except Exception as e:
        failed += len(group)
        print(f"Failed {url}: {e}")

print(f"\n✅ Done! Train: {len(train_features)}, Eval: {len(eval_features)}, Failed: {failed}")

# Save to disk
with open(f'{save_dir}/train.pkl', 'wb') as f:
    pickle.dump(train_features, f)
with open(f'{save_dir}/eval.pkl', 'wb') as f:
    pickle.dump(eval_features, f)
print("✅ Features saved to Drive")
gc.collect()
```

```python
import pickle, gc, torch, os
import numpy as np
from torch.utils.data import Dataset as TorchDataset, DataLoader
from transformers import WhisperProcessor, WhisperForConditionalGeneration, WhisperTokenizer, get_linear_schedule_with_warmup
from torch.optim import AdamW
import evaluate
from torch.cuda.amp import autocast, GradScaler

print(f"GPU: {torch.cuda.get_device_name(0)}")
DEVICE = torch.device('cuda')

# ── Load features ──────────────────────────────────────────────────────────
save_dir = '/content/drive/MyDrive/josh_talks_asr/features'
with open(f'{save_dir}/train.pkl', 'rb') as f:
    train_raw = pickle.load(f)
with open(f'{save_dir}/eval.pkl', 'rb') as f:
    eval_raw = pickle.load(f)
print(f"Train: {len(train_raw)}, Eval: {len(eval_raw)}")

# ── Dataset ────────────────────────────────────────────────────────────────
class WhisperDataset(TorchDataset):
    def __init__(self, features):
        self.features = features
    def __len__(self):
        return len(self.features)
    def __getitem__(self, idx):
        return {
            # float32 this time — fixes NaN loss
            'input_features': torch.tensor(self.features[idx]['input_features'], dtype=torch.float32),
            'labels': self.features[idx]['labels']
        }

def collate_fn(batch):
    input_features = torch.stack([b['input_features'] for b in batch])
    max_len = max(len(b['labels']) for b in batch)
    padded = []
    for b in batch:
        labs = b['labels']
        padded.append(labs + [-100] * (max_len - len(labs)))
    labels = torch.tensor(padded, dtype=torch.long)
    return {'input_features': input_features, 'labels': labels}

train_loader = DataLoader(WhisperDataset(train_raw), batch_size=4, shuffle=True, collate_fn=collate_fn)
eval_loader  = DataLoader(WhisperDataset(eval_raw),  batch_size=4, shuffle=False, collate_fn=collate_fn)
del train_raw, eval_raw
gc.collect()
print("✅ Dataloaders ready")

# ── Model — float32 ────────────────────────────────────────────────────────
MODEL = "openai/whisper-small"
processor = WhisperProcessor.from_pretrained(MODEL, language="hindi", task="transcribe")
tokenizer = WhisperTokenizer.from_pretrained(MODEL, language="hindi", task="transcribe")
# Load in float32 (NOT float16)
model = WhisperForConditionalGeneration.from_pretrained(MODEL)
model.generation_config.language = "hindi"
model.generation_config.task = "transcribe"
model.generation_config.forced_decoder_ids = None
model = model.to(DEVICE)
print("✅ Model loaded in float32")

# ── Optimizer + scaler for mixed precision ─────────────────────────────────
optimizer = AdamW(model.parameters(), lr=1e-5, weight_decay=0.01)
scaler = GradScaler()  # handles fp16 safely during backward pass
wer_metric = evaluate.load("wer")
save_path = '/content/drive/MyDrive/josh_talks_asr/whisper-small-hindi'
os.makedirs(save_path, exist_ok=True)
best_wer = float('inf')

print(f"\n🚀 Training 3 epochs...")

for epoch in range(3):
    model.train()
    total_loss = 0
    valid_steps = 0

    for step, batch in enumerate(train_loader):
        input_features = batch['input_features'].to(DEVICE)
        labels = batch['labels'].to(DEVICE)

        # Mixed precision: autocast for forward, scaler for backward
        with autocast():
            outputs = model(input_features=input_features, labels=labels)
            loss = outputs.loss

        if torch.isnan(loss) or torch.isinf(loss):
            del input_features, labels, outputs, loss
            continue

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad()

        total_loss += loss.item()
        valid_steps += 1

        del input_features, labels, outputs, loss
        if step % 20 == 0:
            avg = total_loss / max(valid_steps, 1)
            print(f"  Epoch {epoch+1} | Step {step} | Loss {avg:.4f}")
            gc.collect()

    print(f"\n📊 Epoch {epoch+1} avg loss: {total_loss/max(valid_steps,1):.4f}")

    # ── Eval ────────────────────────────────────────────────────────────────
    model.eval()
    all_preds, all_refs = [], []
    with torch.no_grad():
        for batch in eval_loader:
            input_features = batch['input_features'].to(DEVICE)
            labels = batch['labels'].clone()
            labels[labels == -100] = tokenizer.pad_token_id
            generated = model.generate(input_features, language="hi", task="transcribe", max_new_tokens=128)
            all_preds.extend(tokenizer.batch_decode(generated, skip_special_tokens=True))
            all_refs.extend(tokenizer.batch_decode(labels, skip_special_tokens=True))
            del input_features, generated
            gc.collect()

    epoch_wer = wer_metric.compute(predictions=all_preds, references=all_refs)
    print(f"📊 Epoch {epoch+1} WER: {epoch_wer*100:.1f}%")

    if epoch_wer < best_wer:
        best_wer = epoch_wer
        model.save_pretrained(save_path)
        processor.save_pretrained(save_path)
        print(f"✅ Best model saved! WER={best_wer*100:.1f}%")

print(f"\n🎉 Done! Best WER: {best_wer*100:.1f}%")
```

```python
# Correct evaluation on Hindi data
from datasets import load_dataset
from transformers import WhisperForConditionalGeneration, WhisperProcessor
import torch, gc, evaluate, pandas as pd

DEVICE = torch.device('cuda')
wer_metric = evaluate.load("wer")

MODEL_PATH = "openai/whisper-small"
FT_PATH = "/content/drive/MyDrive/josh_talks_asr/whisper-small-hindi"

# Load Hindi FLEURS test set
print("Loading Hindi FLEURS...")
fleurs = load_dataset("google/fleurs", "hi_in", split="test")
print(f"✅ {len(fleurs)} Hindi test samples")
print(f"Sample ref: {fleurs[0]['transcription'][:60]}")

def evaluate_model(model, processor, name, max_samples=100):
    model.eval()
    preds, refs = [], []
    for i, sample in enumerate(fleurs):
        if i >= max_samples: break
        try:
            audio = sample["audio"]["array"]
            ref = sample["transcription"]
            inputs = processor(audio, sampling_rate=16000, return_tensors="pt")
            with torch.no_grad():
                generated = model.generate(
                    inputs.input_features.to(DEVICE),
                    language="hi", task="transcribe", max_new_tokens=225
                )
            pred = processor.tokenizer.decode(generated[0], skip_special_tokens=True)
            preds.append(pred)
            refs.append(ref)
            if i % 25 == 0:
                print(f"[{name}] {i}: Ref={ref[:40]} | Pred={pred[:40]}")
            del generated; gc.collect()
        except Exception as e:
            print(f"Skip {i}: {e}")
    return wer_metric.compute(predictions=preds, references=refs), preds, refs

# Baseline
print("\n📊 Baseline Whisper-small on Hindi FLEURS...")
model_base = WhisperForConditionalGeneration.from_pretrained(MODEL_PATH).to(DEVICE)
model_base.generation_config.language = "hindi"
model_base.generation_config.task = "transcribe"
model_base.generation_config.forced_decoder_ids = None
wer_base, preds_base, refs = evaluate_model(model_base, WhisperProcessor.from_pretrained(MODEL_PATH, language="hindi", task="transcribe"), "Base")
print(f"✅ Baseline WER: {wer_base*100:.1f}%")
del model_base; gc.collect(); torch.cuda.empty_cache()

# Fine-tuned
print("\n📊 Fine-tuned model on Hindi FLEURS...")
model_ft = WhisperForConditionalGeneration.from_pretrained(FT_PATH).to(DEVICE)
model_ft.generation_config.language = "hindi"
model_ft.generation_config.task = "transcribe"
model_ft.generation_config.forced_decoder_ids = None
wer_ft, preds_ft, _ = evaluate_model(model_ft, WhisperProcessor.from_pretrained(FT_PATH, language="hindi", task="transcribe"), "FT")
print(f"✅ Fine-tuned WER: {wer_ft*100:.1f}%")

# Results table
print("\n" + "="*55)
print("WER TABLE — FLEURS Hindi Test Set (n=100)")
print("="*55)
print(f"{'Model':<38} {'WER':>8}")
print("-"*55)
print(f"{'Whisper-small (Pretrained)':<38} {wer_base*100:>7.1f}%")
print(f"{'Whisper-small (Fine-tuned, Josh Talks)':<38} {wer_ft*100:>7.1f}%")
print(f"{'Improvement':<38} {(wer_base-wer_ft)*100:>+7.1f}%")
print("="*55)

# Save for error analysis
df = pd.DataFrame({'reference': refs, 'baseline': preds_base, 'finetuned': preds_ft})
df.to_csv('/content/drive/MyDrive/josh_talks_asr/hindi_fleurs_results.csv', index=False)
print(f"✅ Saved results")
```
